```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("yellow_tripdata_2021-01.csv", nrows=100, index_col=0)
display(df)
```

```python
print(pd.io.sql.get_schema(df, name = "yellow_taxi_data"))
#Creating table but need to transform date into dateformat
```

```python
df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])
print(pd.io.sql.get_schema(df, name = "yellow_taxi_data"))
```

```python
for i, row in df.iterrows():
    print("data :", i, row , "\n")
```

```python
engine = create_engine("postgresql://root:root@localhost:5432/ny_taxi")
#postgresql. user:pswd localhost:port/dbname
```

```python
print(pd.io.sql.get_schema(df, name = "yellow_taxi_data",con = engine ))
#Saved as a postgre schema
```

```python
### since the file is too big we are going to use an iterator
df_iter = pd.read_csv("yellow_tripdata_2021-01.csv",iterator=True, chunksize=100000,index_col=0 )
```

```python
from time import time
```

```python
while True:
    t_start = time()
    df=next(df_iter)
    df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
    df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])
    df.to_sql(name = "yellow_taxi_data", con=engine, if_exists="append")
    t_end=time()
    print("Inserted another chunk, it took : %3.f seconds" %(t_end-t_start))
    
```
